```python
import logging
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
import pandas as pd
import random
import time

current_time = datetime.now().strftime('%Y-%m-%d_%H-%M-%S')
log_filename = f"{current_time}_AmazonScraper.log"

# Configure logging to store logs in a file
logging.basicConfig(filename=log_filename, level=logging.INFO, format='%(asctime)s - %(levelname)s: %(message)s')

# Create a logger
logger = logging.getLogger()

# Create a console handler and set the level to INFO
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.INFO)

# Create a file handler and set the level to INFO
file_handler = logging.FileHandler(log_filename)  # Use the dynamic log filename
file_handler.setLevel(logging.INFO)

# Create a formatter and attach it to the handlers
formatter = logging.Formatter('%(asctime)s - %(levelname)s: %(message)s')
console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)

# Add the handlers to the logger
logger.addHandler(console_handler)
logger.addHandler(file_handler)
```

```python
# Initialize the Selenium WebDriver
from  selenium import webdriver


#Specify the path to the custom Chrome binary
chrome_binay_path = r"C:\Users\rajesh.hugar\Downloads\Amazon_Webscraping\chrome-win64\chrome-win64\chrome.exe"

# Create Chrome options and set the binary location
chrome_options = webdriver.ChromeOptions()
chrome_options.binary_location= chrome_binay_path

# Create a Service object for ChromeDriver

chromedriver_path = r"C:\Users\rajesh.hugar\Downloads\Amazon_Webscraping\chromedriver-win64\chromedriver-win64\chromedriver.exe"
service = webdriver.chrome.service.Service(chromedriver_path)



#Initiate The Chrome Webdriver With the specified options and service
driver = webdriver.Chrome(service=service,options=chrome_options)
```

```python
# Create an empty DataFrame
df = pd.DataFrame(columns=["Brand", "URL", "Product Name", "Title", "Text", "Posted On", "Stars", "Site", "Scraping_date"])
```

```python
# List of URLs to scrape
urls = [

    "https://www.amazon.com/Dawn-EZ-Squeeze-Dishwashing-Non-Scratch-Scrubber/product-reviews/B09N7TRGX6/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber={i}"
]

for url in urls:
    for i in range(1, 13):  # Scrape 12 pages per URL
        try:
            logging.info(f"Scraping {url} - Page {i}")
            formatted_url = url.format(i=i)

            driver.get(formatted_url)
            brand = "Dawn"
            product_name = WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.XPATH, "//a[@data-hook='product-link']"))
            )
            logging.info(product_name.text)
            sleep_time = random.randint(5, 15)
            time.sleep(sleep_time)

            text = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//div[@id='cm_cr-review_list']//span[@data-hook='review-body']"))
            )
            title = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located(
                    (By.XPATH, "//div[@id='cm_cr-review_list']//a[@data-hook='review-title']//span[not(@*)] | //div[@id='cm_cr-review_list']//a[@data-hook='review-title']//span[@class='cr-original-review-content']")
                )
            )
            logging.info(f'Getting data from page {i}')

            dates = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//div[@id='cm_cr-review_list']//span[@data-hook='review-date']"))
            )
            stars = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.XPATH, "//div[@id='cm_cr-review_list']//i[@data-hook='review-star-rating']//span"))
            )

            for j in range(len(dates)):
                df = df._append(
                    {
                        "Brand": brand,
                        "URL": formatted_url,
                        "Product Name": product_name.text,
                        "Title": title[j].text,
                        "Text": text[j].text,
                        "Posted On": dates[j].text,
                        "Stars": stars[j].get_attribute('textContent'),
                        "Site": "Amazon",
                        "Scraping_date": datetime.now()
                    },
                    ignore_index=True
                )

        except Exception as e:
            logging.error(f"Error on page {i}: {str(e)}")
            break

# Close the WebDriver
driver.quit()

# Save the scraped data to a CSV file
df.to_csv("amazon_reviews.csv", index=False)
```
